```python
import numpy as np
import matplotlib.pyplot as plt

def regressao_linear(x, y):
    """
    Calcula a regressão linear pelo método dos mínimos quadrados
    Retorna: a (coeficiente angular), b (coeficiente linear), r2 (coeficiente de determinação)
    Equação: y = a*x + b
    """
    n = len(x)

    # Somatórios necessários
    soma_x = np.sum(x)
    soma_y = np.sum(y)
    soma_xy = np.sum(x * y)
    soma_x2 = np.sum(x**2)

    # Cálculo dos coeficientes
    a = (n * soma_xy - soma_x * soma_y) / (n * soma_x2 - soma_x**2)
    b = (soma_y - a * soma_x) / n

    # Coeficiente de determinação R²
    y_medio = np.mean(y)
    ss_tot = np.sum((y - y_medio)**2)
    ss_res = np.sum((y - (a*x + b))**2)
    r2 = 1 - (ss_res / ss_tot)

    return a, b, r2

def plotar_regressao(x, y, a, b, titulo, xlabel, ylabel):
    """Plota os dados e a reta de regressão"""
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', s=50, label='Dados experimentais')

    # Reta ajustada
    x_linha = np.linspace(min(x), max(x), 100)
    y_linha = a * x_linha + b
    plt.plot(x_linha, y_linha, 'r-', linewidth=2, label=f'y = {a:.4f}x + {b:.4f}')

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(titulo)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

print("=" * 80)
print("REGRESSÃO LINEAR - MÉTODO DOS MÍNIMOS QUADRADOS")
print("=" * 80)

# ==================== QUESTÃO 1 ====================
print("\n1) PRIMEIRA REGRESSÃO LINEAR")
print("-" * 80)
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2.2, 3.3, 4.2, 5.1, 6.3])

a1, b1, r2_1 = regressao_linear(x1, y1)

print(f"Equação da reta: y = {a1:.4f}x + {b1:.4f}")
print(f"Coeficiente de determinação R² = {r2_1:.6f}")
print(f"Coeficiente angular (a) = {a1:.4f}")
print(f"Coeficiente linear (b) = {b1:.4f}")

plotar_regressao(x1, y1, a1, b1,
                 'Questão 1 - Regressão Linear', 'X', 'Y')

# ==================== QUESTÃO 2 ====================
print("\n2) MÉTODO DOS MÍNIMOS QUADRADOS - SEGUNDA TABELA")
print("-" * 80)
x2 = np.array([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.6])
y2 = np.array([-1.1, -1.9, -2.9, -3.9, -4.8, -6.0, -6.8])

a2, b2, r2_2 = regressao_linear(x2, y2)

print(f"Equação da reta: y = {a2:.4f}x + ({b2:.4f})")
print(f"Coeficiente de determinação R² = {r2_2:.6f}")
print(f"Coeficiente angular (a) = {a2:.4f}")
print(f"Coeficiente linear (b) = {b2:.4f}")

plotar_regressao(x2, y2, a2, b2,
                 'Questão 2 - Método dos Mínimos Quadrados', 'X', 'Y')

# ==================== QUESTÃO 3 ====================
print("\n3) BRILHO DA ESTRELA AO LONGO DOS ANOS")
print("-" * 80)
anos = np.array([2003, 2004, 2005, 2006, 2007])
brilho = np.array([3.2, 3.0, 2.9, 2.85, 2.7])

# Para facilitar, usar anos relativos (0, 1, 2, 3, 4)
anos_relativos = anos - 2003

a3, b3, r2_3 = regressao_linear(anos_relativos, brilho)

print(f"Usando anos relativos a 2003:")
print(f"Equação da reta: Brilho = {a3:.4f} * (Ano - 2003) + {b3:.4f}")
print(f"Ou: Brilho = {a3:.4f} * Ano + {b3 - a3*2003:.4f}")
print(f"Coeficiente de determinação R² = {r2_3:.6f}")

# Previsões
print(f"\nPrevisões:")
print(f"Brilho em 2008: {a3 * 5 + b3:.4f}")
print(f"Brilho em 2010: {a3 * 7 + b3:.4f}")

plotar_regressao(anos, brilho, a3, b3 + a3*2003,
                 'Questão 3 - Brilho da Estrela', 'Ano', 'Brilho')

# ==================== QUESTÃO 4 ====================
print("\n4) ELONGAÇÃO PERCENTUAL vs TEMPERATURA")
print("-" * 80)
temperatura = np.array([200, 250, 300, 375, 425, 475, 600])
elongacao = np.array([7.5, 8.6, 8.7, 10, 11.3, 12.7, 15.3])

a4, b4, r2_4 = regressao_linear(temperatura, elongacao)

print(f"Equação da reta: Elongação = {a4:.6f} * Temperatura + {b4:.4f}")
print(f"Coeficiente de determinação R² = {r2_4:.6f}")

# Previsão para 400°C
temp_previsao = 400
elong_previsao = a4 * temp_previsao + b4
print(f"\nPrevisão para 400°C:")
print(f"Elongação percentual = {elong_previsao:.2f}%")

plotar_regressao(temperatura, elongacao, a4, b4,
                 'Questão 4 - Elongação vs Temperatura',
                 'Temperatura (°C)', 'Elongação (%)')

# ==================== QUESTÃO 5 ====================
print("\n5) ÁREA SUPERFICIAL vs PESO")
print("-" * 80)
peso = np.array([70, 75, 77, 80, 82, 84, 87, 90])
area = np.array([2.10, 2.12, 2.15, 2.20, 2.22, 2.23, 2.26, 2.30])

a5, b5, r2_5 = regressao_linear(peso, area)

print(f"Equação da reta: A = {a5:.6f} * W + {b5:.4f}")
print(f"Coeficiente angular (a) = {a5:.6f}")
print(f"Coeficiente linear (b) = {b5:.4f}")
print(f"Coeficiente de determinação R² = {r2_5:.6f}")

# Previsão para 95 kg
peso_previsao = 95
area_previsao = a5 * peso_previsao + b5
print(f"\nPrevisão para pessoa de 95 kg:")
print(f"Área superficial = {area_previsao:.4f} m²")

plotar_regressao(peso, area, a5, b5,
                 'Questão 5 - Área Superficial vs Peso',
                 'Peso (kg)', 'Área (m²)')

# ==================== QUESTÃO 6 ====================
print("\n6) LARGURA DA FAIXA vs DISTÂNCIA (BICICLETAS)")
print("-" * 80)
distancia = np.array([2.4, 2.5, 2.4, 1.8, 1.8, 2.9, 1.2, 3.0, 1.2])
largura = np.array([2.9, 2.1, 2.3, 2.1, 1.8, 2.7, 1.5, 2.9, 1.5])

# Regressão: Distância em função da Largura
a6, b6, r2_6 = regressao_linear(largura, distancia)

print(f"Equação da reta: Distância = {a6:.4f} * Largura + {b6:.4f}")
print(f"Coeficiente de determinação R² = {r2_6:.6f}")

# Largura mínima para distância de segurança de 1,8 m
dist_seguranca = 1.8
largura_minima = (dist_seguranca - b6) / a6

print(f"\nPara distância de segurança mínima de {dist_seguranca} m:")
print(f"Largura mínima da faixa = {largura_minima:.4f} m")

plotar_regressao(largura, distancia, a6, b6,
                 'Questão 6 - Distância vs Largura da Faixa',
                 'Largura da Faixa (m)', 'Distância (m)')
```
